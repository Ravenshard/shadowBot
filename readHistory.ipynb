```python
import discord
import msgClass
import re, os, glob
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
path = "./log/Dominion (Card Game)/border-village"
dirs = glob.glob("{}/*/*".format(path))
for file in dirs:
    df = pd.read_csv(file, delimiter="\t")
    messagesSent = len(df["Content"])
    charCount = 0
    for x in df["Content"]: charCount += len(str(x))
    print("File: {}".format(file.split("/")[4]))
    print("Messages: {}".format(messagesSent))
    print("Chars Sent: {}".format(charCount))
    print("Char per Message: {}\n".format(charCount/messagesSent))
```
